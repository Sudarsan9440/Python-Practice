```python
people ={
    "first": ["John","Jane","Joe"],
    "last": ["Doe","Doe","Doe"],
    "email": ["john.doe@example.com","jane.doe@example.com","joe.doe@example.com"]
}
```

```python
import pandas as pd
```

```python
df=pd.DataFrame(people)
```

```python
filt =(df['last'] == 'Doe') & (df['first'] =='John')
```

```python
df.loc[filt,['first', 'email']]
```

```python
fil_or= (df['email'].str.contains('john')) | (df['first'].str.contains('Joe'))
```

```python
df.loc[fil_or,['first','last']]
```

```python
df.loc[~fil_or,['first','last']] # opposite
```
