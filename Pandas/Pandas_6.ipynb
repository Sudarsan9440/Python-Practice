```python
import pandas as pd
```

```python
people ={
    'first':["Jane", "John", "James"],
    'last':["Doe", "Doe", "Doe"],
    'email':["jane@example.com", "john@example.com", "james@example.com"]
}
```

```python
df=pd.DataFrame(people)
```

```python
df['full_name']=df['first'] + ' ' + df['last']
```

```python
df
```

```python
df.drop(columns=['first', 'last'], inplace=True)
```

```python
df
```

```python
df['full_name'].str.split(' ')
```

```python
df[['first' ,'last']]=df['full_name'].str.split(' ',expand=True)
df
```

```python
df
```

```python
df.append({"first":"Tony"})
```

```python
df.append({"first": "Tony"},inplace=True)
```

```python
# Use pd.concat to add a new row as a DataFrame
new_row = pd.DataFrame({"first": ["Tony"]})
df = pd.concat([df, new_row], ignore_index=True)
df
```

```python
df.drop_duplicates(subset=['first'], keep='first')
```
