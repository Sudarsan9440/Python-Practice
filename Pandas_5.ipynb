```python
import pandas as pd
```

```python
people ={
    "first": ["John","Jane","Joe"],
    "last": ["Doe","Doe","Doe"],
    "email": ["john.doe@example.com","jane.doe@example.com","joe.doe@example.com"]
}
```

```python
df=pd.DataFrame(people)
df
```

```python
df.columns=['first name', 'last name', 'email']
df
```

```python
df.columns=df.columns.str.replace(" ","_")
df
```
